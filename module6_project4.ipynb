```python
#An inner merge operation between the orders table and products table.
order_product = pd.merge(orders, products, how="inner", on=["product_id"])
order_product.head()
```

**OBSERVATION**
The inner merge conbines the columns' elements that are common between the orders and products table

```python
#Use the resulting table to perform another inner merge operation with the customers table.
customer_order_product = pd.merge(order_product, customers, how="inner")
customer_order_product.head()
```

**OBSERVATION**
: The inner merge add the columns' elements that are common in the merge of orders/products table and customers table.

```python
#Perform an inner merge operation between customers and orders tables. 
#Here you need to specify the columns you’ll be merging on explicitly (see the .merge function arguments ‘left_on’, ‘right_on’ and ‘suffixes’).
customer_order = pd.merge(orders, customers, how="inner", left_on=["quantity"], right_index = True)
customer_order
```

**OBSERVATION**
: The merge only combines "quantity" in the orders table that is corresponding to the index of the customers table
