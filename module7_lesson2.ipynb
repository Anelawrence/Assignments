```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt

df = pd.read_csv('anscombe_quartet_data.csv')
df.head()
```

```python
df.describe()
```

```python
df.corr()
```

```python
x = [0,15]
y = [0,15]

x1 = df['x1']
y1 = df['y1']
plt.subplot(2, 2, 1)
plt.title('(x1,y1)')
plt.scatter(x1, y1, c ="blue")
plt.plot(x,y, color = 'black', linestyle = '-')


x2 = df['x2']
y2 = df['y2']
plt.subplot(2, 2, 2)
plt.title('(x2,y2)')
plt.scatter(x2, y2, c ="purple")
plt.plot(x,y, color = 'black', linestyle = '-')

x3 = df['x3']
y3 = df['y3']
plt.subplot(2, 2, 3)
plt.title('(x3,y3)')
plt.scatter(x3, y3, c ="red")
plt.plot(x,y, color = 'black', linestyle = '-')

x = [0,20]
y = [0,20]
x4 = df['x4']
y4 = df['y4']
plt.subplot(2, 2, 4)
plt.title('(x4,y4)')
plt.scatter(x4, y4, c ="green")
plt.plot(x,y, color = 'black', linestyle = '-')

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
plt.show
```

```python
x = [0,15]
y = [0,15]

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

ax1.scatter(x1, y1, c ="blue")
ax1.plot(x, y, color='black')
ax1.set_title('(x1,y1)')

ax2.scatter(x2, y2, c ="purple")
ax2.plot(x, y, color='black')
ax2.set_title('(x2,y2)')

ax3.scatter(x3, y3, c ="red")
ax3.plot(x, y, color='black')
ax3.set_title('(x3,y3)')

x = [0,20]
y = [0,20]
ax4.scatter(x4, y4, c ="green")
ax4.plot(x, y, color='black')
ax4.set_title('(x4,y4)')

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
```

```python
x = [0,15]
y = [0,15]

fig = plt.figure(figsize=(8, 6))

ax5 = fig.add_subplot(221, title='Plot 1')
ax5.scatter(x1, y1, c ="blue")
ax5.plot(x, y, color='green')

ax6 = fig.add_subplot(222, title='Plot 2')
ax6.scatter(x2, y2, c ="purple")
ax6.plot(x, y, color='red')

ax7 = fig.add_subplot(223, title='Plot 3')
ax7.scatter(x3, y3, c ="red")
ax7.plot(x, y, color='black')

x = [0,20]
y = [0,20]
ax8 = fig.add_subplot(224, title='Plot 4')
ax8.scatter(x4, y4, c ="green")
ax8.plot(x, y, color='purple')

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
```
