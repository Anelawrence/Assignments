```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt

sales = pd.read_csv('company_sales_data.csv')
sales.head()
```

```python
x = sales['month_number']
y1 = sales['facecream']
plt.plot(x,y1, color = 'purple', linestyle = '--', marker = '8')

x = sales['month_number']
y2 = sales['toothpaste']
plt.plot(x,y2, color = 'blue', linestyle = '-.', marker = 'o')

x = sales['month_number']
y3 = sales['facewash']
plt.plot(x,y3, color = 'red', linestyle = ':')

x = sales['month_number']
y4 = sales['bathingsoap']
plt.plot(x,y4, color = 'green', linestyle = 'solid', marker = 'o')

x = sales['month_number']
y5 = sales['shampoo']
plt.plot(x,y5, color = 'yellow', linestyle = '-', marker = '+')

x = sales['month_number']
y6 = sales['moisturizer']
plt.plot(x,y6, color = 'black', linestyle = 'None', marker = '*')

plt.xlabel('Months in number')
plt.ylabel('Sales')
plt.title('MONTHLY SALES')
plt.legend(['facecream', 'toothpaste', 'facewash', 'bathingsoap', 'shampoo', 'moisturizer'], loc=2)
plt.show()
```
